# Battle of the neighbourhoods - Where to live in London

## Problem Description

According to statistics provided by [Statista.com](https://www.statista.com/statistics/1029293/house-sales-volume-in-london/), tens of thousands of people purchase properties in London each year and will continue to do so even when net population for the city is in decline. The challenge faced by most of these people is determining in which area of London they would like to live given the variety present here as in most global cities.  

A number of people may have a location in mind, but then find that the prices of houses in that area are too high or another element is missing. We will therefore group areas of a similar style in terms of amenities and healthcare access sourced from [Foursquare](https://api.foursquare.com), and then provide supplementary data such as house prices, transport links and crime statistics to help them decide on possible alternatives.  Those without a starting point in mind will still benefit from a summary of the types of areas within London. 


## Data requirements and usage

The initial data set of borough names will be enhanced using Geopy data.  Once this has been cleaned it will be populated with venue data from Foursquare and then grouped into similar areas based on these venues.  Details on house prices, crime levels and transport links will then be added to the data set to enable further scoring of desirable places to live.

The non-Foursquare data sources will all be sourced from government institutions as this is a validated source of data with consistent nomenclature and classification.  This should reduce the manipulation required to integrate the data sets and reduce our risk of drawing false conclusions from the data.

* Boroughs and post-codes
    - London Data Store
    - Used to source an accurate list of boroughs within London.
* Median House Prices
    - [Land Registry House Prices](https://data.london.gov.uk/download/average-house-prices/b1b0079e-698c-4c0b-b8c7-aa6189590ca4/land-registry-house-prices-borough.csv)
    - Median house price per year per borough on a quarterly frequency which will allow us to calculate current prices and indicative growth over a suitable period.
    - Simple data set consisting of an area identifier, area, price and period
* Geolocation data
    - [Geopy](https://pypi.org/project/geopy/)
    - Used to return the geolocation data based on the borough names/addresses
    - Latitude and longitude for our boroughs will be sourced here
* Social Amenities 
    - [Foursquare](https://api.foursquare.com)
    - The venues data will be sourced from here via the Foursquare API
* Access to healthcare
    - [NHS Hospitals](http://media.nhschoices.nhs.uk/data/foi/Hospital.csv)

    - Contains hospital names, location data and contact details
* Access to public transport
     - [Tube locations](https://www.whatdotheyknow.com/request/512947/response/1238210/attach/3/Stations%2020180921.csv.txt)
     - Contains tube names, latitude and longitude, along with the tube lines and zones which may be useful supplemental data fir extensions
* Crime statistics
    - [Met Police Crime Dashboard](https://www.met.police.uk/sd/stats-and-data/met/crime-data-dashboard/)
    - Number of crimes committed per borough broken down by type back to April 2010 which will allow for trend analysis as well as crime type for scoring
    - Simple CSV exports available to be keyed on borough
